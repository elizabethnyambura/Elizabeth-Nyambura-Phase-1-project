```python
#find the total gross generated
total_gross_income = (joined_imdb_df1['domestic_gross'] + joined_imdb_df1['foreign_gross'])
total_gross_income.value_counts()
```

```python
#create total gross income column
joined_imdb_df1 = joined_imdb_df1.eval('total_gross_income = domestic_gross + foreign_gross')
joined_imdb_df1.head()
```

DATA VISUALIZATION

 relationship between genres and total gross income

```python
most_profitable_genres = joined_imdb_df1.groupby('genres')['total_gross_income'].sum()[:10]
most_profitable_genres = most_profitable_genres.sort_values(ascending=False)
most_profitable_genres
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

width = 0.4

most_profitable_genres.plot(kind='bar', color='blue', ax=ax, width=width, position=1)

ax.legend()
plt.xlabel('Top ten Profitable Genres')
plt.ylabel('Frequency')
plt.show()
```

As per the above graph, Action adventureFantasy have the highest gross income while Action Adventure have the least gross income.

Relationship between Genres and Studio

```python
most_profitable_studios = joined_imdb_df1.groupby('studio')['total_gross_income'].sum()[:10]
most_profitable_studios
```

```python
#Plot a bar graph to view and analyze the distribution and determine the best studio to invest in

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)

width = 0.5

most_profitable_studios.plot(kind='bar', color= 'green', ax=ax, width=width, position=1)
plt.xlabel('Studio', fontsize = 16)
plt.ylabel('Frequency', fontsize = 16)
ax.set_title("most Profitable Studios", fontsize = 16)

ax.legend()
plt.show()
```

As per the above graph,A24 studio has the highest returns in terms of total gross income while 3D studio and AGF studio has the least and no returns respectively.It is therefore advisable for Microsoft to invest in the A24 studio

movies Genre with the highest average rating

```python
genres_rating = joined_imdb_df1.groupby('genres')['averagerating'].mean().sort_values(ascending = False)
genres_rating.head()
```

```python
#bar graph showing the average rating by genre
fig, ax =plt.subplots(figsize=(14,6))
order_by = joined_imdb_df1.groupby('genres')['averagerating'].mean().sort_values(ascending=False).index
sns.barplot(data=joined_imdb_df1, x='genres', y= 'averagerating', palette='muted', order=order_by, ax=ax)
ax.set_title('Average rating by genre')
ax.set_ylabel('average rating')
plt.xticks(rotation=60);
```

As per the above analysis,Adventure, action,Sport and Adventure,Drama,Sci-Fi have the highest ratings while Documentary,Sport have the lowest rating  

Relationship between the average rating and the gross income

```python
#scatter plot showing  total gross vs average rating
fig, ax =plt.subplots(figsize=(11,6))
sns.scatterplot(x=joined_imdb_df1['averagerating'], y=joined_imdb_df1['total_gross_income'], ax=ax)
ax.set_title('total gross vs average rating')
ax.set_ylabel('total gross income')
ax.set_xlabel('average rating');
```

As per the above analysis,there is no clear relationship between average rating and total gross rating and therefore this relationship should not be used for decision making on the business investment

Relationship between Gross income and start year

```python
#lineplot showing the domestic gross and foreign gross income by start year
fig, ax =plt.subplots(figsize=(15,7))
sns.lineplot(data=joined_imdb_df1, x='start_year', y='domestic_gross', ax=ax, label='domestic_gross')
sns.lineplot(data=joined_imdb_df1, x='start_year', y='foreign_gross', label='foreign_gross', ax=ax)
ax.set_title('total_gross_income by start year')
ax.set_ylabel('total_gross_income');
```

The foreign gross income is constantly higher than the domestic gross income throughout the years when a movie is released.
Movies that were released in 2018 had the highest foreign gross and domestic gross income.The gross incomes then drastically dropped to their lowest for movies that were released in 2019.
